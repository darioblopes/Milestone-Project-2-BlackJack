```python
# imported the random libary 

import random

# set up global variables using tuples to represent the suits and ranks of the Card class
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')

# set up a Dictionary values global variable Dictionary to define the values of all the cards. 
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
        'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        
        return self.rank+ " of "+self.suit # way of returning the card name as string
        
```

```python
class Deck:
    
    def __init__(self):
        """
        Using the card class to quickly create 52 cards in the deck using a nested for-loop
        """
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                
                # self.deck is list of the card objects
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        """
        Way to show all the cards in the deck in string representation 
        """
        deck_comp = ''
        for card in self.deck:
            deck_comp += '\n'+ card.__str__()
        return "The deck has: "+deck_comp
    

    def shuffle(self):
        """
        Shuffling the list in place. 
        """
        random.shuffle(self.deck)
        
    def deal(self):
        """
        Dealing the cards by popping off one card from the deck and returning that card value
        """
        single_card = self.deck.pop()
        return single_card
```

```python
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        # card passed in
        # from Deck.deal() --> single(suit,rank)
        self.cards.append(card)
        self.value += values[card.rank]
        
        # track aces
        if card.rank == "Ace":
            self.aces +=1
        
    
    def adjust_for_ace(self):
        
        # IF TOTAL VALUE > 21 AND I STILL HAVE AN ACE
        # THAN CHANGE MY ACE TO BE A 1 INSTEAD OF AN 11
        while self.value > 21 and self.aces > 0:
            self.value -= 10 
            self.aces -= 1 
```

```python
class Chips:
    
    def __init__(self):
        self.total = 100  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    """
    Getting a input from the user and setting the current bet
    """
    
    while True:
        
        try:
            chips.bet = int(input("How many chips would you like to bet? "))
        except:
            print("Sorry please provide an integer")
            
        else:
            if chips.bet > chips.total:
                print("Sorry, you do not have enough chips! You have: {}".format(chips.total))
            else:
                break
```

```python
def hit(deck,hand):
    """
    Function for the player to take a hit (new card)
    """
    
    single_card = deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    """
    Function that will ask the user if it wants to stand with its current hand or take a hit (new card)
    IF they didn't provide a h or s it will contuine 
    """
    global playing  # to control an upcoming while loop
    
    while True:
        x = input("Hit or Stand? Enter h or s ")
        
        if x[0].lower() == "h":
            hit(deck,hand)
            
        elif x[0].lower() == "s":
            playing = False
            
        else:
            print("Sorry, I did not understand that, Please enter h or s only!")
            continue
            
        break
```

```python
# Two funtions to display the hands

def show_some(player,dealer):
    
    # Show only One of the dealer's cards
    print("\n Dealer's Hand: ")
    print("First card hidden!")
    print(dealer.cards[1])
    
    # Show all (2 cards) od the player's hands/cards
    print("\n Player's Hand:")
    for card in player.cards:
        print(card)
    
def show_all(player,dealer):
    
    
    # Show all the dealer's cards
    print("\n Player's Hand:")
    for card in dealer.cards:
        print(card)
        
    # Calculate and display value (J+K == 20)
    print(f"Value of Dealer's hand is: {dealer.value}")
    
    # Show all the players cards
    print("\n Player's Hand:")
    for card in player.cards:
        print(card)
    print(f"Value of Player's hand is: {player.value}")
```

```python
# Functions to display each of the game situations 

def player_busts(player,dealer,chips):
    print("BUST PLAYER!")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("PLAYER WINS!")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("PLAYER WINS! DEALER BUSTED!")
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print("DEALER WINS")
    chips.lose_bet()
    
    
def push(player,dealer):
    print("Dealer and player tie! PUSH!")
```

```python
while True:
    # Print an opening statement while True
    
    print("WELCOME TO BLACKJACK")

    
    # Create & shuffle the deck, deal two cards to each player
    deck = Deck()
    deck.shuffle()
    
    # Create hand object to represent the player
    player_hand = Hand()
    # add two cards to the player's hand object
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    
    # Create hand object to represent the dealer's hand
    dealer_hand = Hand()
    # add two cards to the dealer's hand
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    
        
    # Set up the Player's chips
    player_chips = Chips()
    
    
    # Prompt the Player for their bet
    take_bet(player_chips)
    
    # Show cards (but keep one dealer card hidden)
    show_some(player_hand,dealer_hand)
    
    while playing:  # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(deck,player_hand)
        
        
        # Show cards (but keep one dealer card hidden)
        show_some(player_hand,dealer_hand)
 
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)
        

            break

    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
    if player_hand.value <= 21:
        
        while dealer_hand.value < 17:
            hit(deck,dealer_hand)
            
    
        # Show all cards
        show_all(player_hand,dealer_hand)
    
        # Run different winning scenarios
        
        # Check if dealer busted
        if dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,player_chips)
        # Dealer wins
        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chips)
        # Player wins
        elif dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,player_chips)
        # Its a push (tie)
        else:
            push(player_hand,dealer_hand)
        
    
    # Inform Player of their chips total 
    print("\n Player total chips are at: {}".format(player_chips.total))
    # Ask to play again
    new_game = input("Would you like to play another hand? y or n")
    
    # loop to reset the playing to true
    if new_game[0] == "y":
        playing = True
        
    # loop to break out the game and print thanks for playing    
    else:
        print("Thank for playing Blackjack!")
        break
```
